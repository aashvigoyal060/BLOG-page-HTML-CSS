```python
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow

model= YOLO('yolov8n.pt')

video_path= '/content/GreenBird.mp4'

cap=cv2.VideoCapture(video_path)

frame_count=0
max_frames = 30

while cap.isOpened() and frame_count < max_frames:
  ret, frame = cap.read()
  if not ret:
    break

  results=model(frame)
  annotated_frame = results[0].plot()

  cv2_imshow(annotated_frame)

  frame_count += 1

cap.release()
```

```python
output_path='/content/output_annonated_video.avi'
fps=cap.get(cv2.CAP_PROP_FPS)
width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

video_path=  '/content/GreenBird.mp4'
cap = cv2.VideoCapture(video_path)
frame_count =0

fourcc = cv2.VideoWriter_fourcc(*'XVID')
out= cv2.VideoWriter(output_path, fourcc, fps, (width, height))

while cap.isOpened() and frame_count < max_frames :
  ret , frame= cap.read()
  if not ret:
    break

  results = model(frame)
  annotated_frame= results[0].plot()

  out.write(annotated_frame)
  cv2_imshow(annotated_frame)

  frame_count += 1

cap.release()

out.release()

print("video saved at:", output_path)
```

```python
# extracrting bounding boxes and class labels
for results in results:
  boxes = results.boxes
  for box in boxes:
    cls_id = int(box.cls[0])
    label=model.names[cls_id]
    conf=float(box.conf[0])

    print(f"Detected:{label}, Confidence: {conf:.2f}")
```

```python
model.predict(source=video_path, save=True, conf=0.3)
```

```python
import pandas as pd
data=[]
for result in results:
  for box in result.boxes:
    cls_id=int(box.cls[0])
    label=model.names[cls_id]
    conf=float(box.conf[0])

    x1, y1, x2, y2 = box.xyxy[0].tolist()

    data.append([label,conf, x1,y1,x2,y2])
df= pd.DataFrame(data, columns=["Label", "Confidence", "X1", "Y1", "X2","Y2"])
df.to_csv("/content/detections.csv", index=False)

print("Saved detection info to CSV")
```
